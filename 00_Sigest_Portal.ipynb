#Importar o ficheiro Transportes

```python

import os
import pandas as pd

# Caminho da pasta 
folder_path = os.path.join(os.path.expanduser("~"), "Desktop", "New folder")

# Obter o primeiro ficheiro encontrado na pasta
files = os.listdir(folder_path)
if not files:
    raise FileNotFoundError("Nenhum ficheiro encontrado na pasta.")

file_path = os.path.join(folder_path, files[0])
print(f"A abrir ficheiro: {file_path}")

ext = os.path.splitext(file_path)[1].lower()

if ext in [".csv", ".txt"]:
    df_transportes = pd.read_csv(file_path)
elif ext in [".xls", ".xlsx"]:
    df_transportes = pd.read_excel(file_path)
else:
    raise ValueError(f"Formato de ficheiro não suportado: {ext}")

```
