```python
import osmnx as ox
import networkx as nx
import folium

# Get the road network graph for Lisbon, Portugal
G = ox.graph_from_place('Lisbon, Portugal', network_type='drive')

# Add edge speeds and travel times to the graph
G = ox.speed.add_edge_speeds(G)
G = ox.speed.add_edge_travel_times(G)

# Find the nearest nodes to specified coordinates
orig = ox.nearest_nodes(G, X=38.77265264723505 , Y=-9.096639520569637)
print("Orig Node:", orig)

dest = ox.nearest_nodes(G, X=38.75669121077172, Y=-9.10432136718318)
print("Dest Node:", dest)

# Check if there is a valid path between the origin and destination
if nx.has_path(G, orig, dest):
    # Find the shortest path from the origin to the destination
    route = nx.shortest_path(G, orig, dest, weight='travel_time')

    # Create a Folium map centered around the origin
    map_center = [38.77265264723505, -9.096639520569637]
    mymap = folium.Map(location=map_center, zoom_start=14)

    # Convert node coordinates to a list of latitudes and longitudes
    route_coordinates = [(G.nodes[node]['y'], G.nodes[node]['x']) for node in route]

    # Add a PolyLine to the map to represent the route
    folium.PolyLine(locations=route_coordinates, color='blue').add_to(mymap)

    # Display the map
    mymap
else:
    print("No valid path between the origin and destination nodes.")
```
