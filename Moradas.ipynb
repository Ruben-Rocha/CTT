```python
import time
import time, psutil, os
start = time.time()
m0 = psutil.Process(os.getpid()).memory_info().rss / 1024**2
```

```python
import pandas as pd

ficheiro = r"C:\Users\1420844\Downloads\addressGeoLoc_export_20221006.csv"

df = pd.read_csv(ficheiro, sep=",", quotechar='"', skiprows=1,
                  names=["MORADA", "CP", "LOCALIDADE", "LATITUDE", "LONGITUDE"])

```

```python
df.head()
```

```python
df.describe()
```

```python
# garantir que a coluna CP é string
df["CP"] = df["CP"].astype(str)

# filtrar e contar
contagem = df[df["CP"].str[:4].isin(["4830", "4850"])]["MORADA"].count()

print("Número de moradas:", contagem)
```

```python
df_filtrado = df[df["CP"].str[:4].isin(["4830", "4850"])]
df_filtrado
```

```python
# ==========================
# 1. Filtrar colunas e CP
# ==========================
df_filtrado = df.loc[
    df["CP"].str[:4].isin(["4830", "4850"]),
    ["MORADA", "CP", "LOCALIDADE"]
]

df_filtrado
```

```python
# ==========================
# 2. Exportar para Excel com colunas definidas
# ==========================
# Lista de colunas finais
colunas_final = ["MORADA", "CP", "LOCALIDADE"]

# Criar DataFrame final (colunas extra ficam a NaN)
df_final = df_filtrado.reindex(columns=colunas_final)

# Caminho da pasta Downloads do utilizador
downloads = os.path.join(os.path.expanduser("~"), "Downloads")

# Nome do ficheiro
ARQ_OUT = os.path.join(downloads, "Moradas_4830_4850.xlsx")

# Exportar
df_final.to_excel(ARQ_OUT, index=False)

```

```python
m1 = psutil.Process(os.getpid()).memory_info().rss / 1024**2
print("Tempo Pandas:", round(time.time() - start, 2), "segundos")
print("Memória Pandas:", round(m1 - m0, 2), "MB")
```
